```python
from crewai import Agent, Task, Crew, Process
from langchain_groq import ChatGroq
from langchain.tools import DuckDuckGoSearchRun
```

```python
from crewai_tools import tool

@tool('DuckDuckGoSearch')
def search(search_query: str):
    """Search the web for information on a given topic"""
    return DuckDuckGoSearchRun().run(search_query)
```

```python
search_tool = search
```

```python
llm = ChatGroq(temperature = 0.5, groq_api_key = "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",model = "llama3-8b-8192")
```

```python
researcher_agent = Agent(
    role = "Senior Researcher",
    goal = "Discover the information",
    backstory = "A curious researcher who always finds the available information across the internet",
    verbose = True,
    tools = [search_tool],
    allow_delegation = False,
    llm = llm
)
```

```python
insight_researcher_writer = Agent(
    role = "Insight Researcher and writer",
    goal = "Discover relevent information like profile, linkedin ID, position, company, etc,",
    backstory = "You are good at finding the relevent information about the person and does not confuses the information with the person of similar name. Check out the company and name so that you wont get confused and return the insights like person profile, company, position, linkedin id,etc. And write the discovered insights",
    verbose = True,
    allow_delegation = False,
    llm = llm
)
```

```python
research_task = Task(
    description='Research about the person "Dev Patel Verchool" across the internet',
    agent=researcher_agent,
    expected_output = "Informations from internet about the person Dev Patel Verchool",
)
```

```python
writer_task = Task(
    description = "Write the insights from the data and dont confuse with similar named persons and write only the relevent indformation ",
    agent = insight_researcher_writer,
    expected_output = "A final markdown format which contains insights about Dev Patel Verchool and only the relevent information.The report must contain person profile info, linkedin id, others , etc ",
)
```

```python
crew = Crew(
    agents = [researcher_agent, insight_researcher_writer],
    tasks = [research_task, writer_task],
    process = Process.sequential
)
```

```python
result = crew.kickoff()
print(result)
```
